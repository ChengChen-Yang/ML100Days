```python
app_train.describe()
```

```python
int_features = []
float_features = []
object_features = []

for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype == 'int64':
        int_features.append(feature)
    elif dtype == 'float64':
        float_features.append(feature)
    else:
        object_features.append(feature)

print(int_features,'\n ============')
print(float_features,'\n ============')
print(object_features,'\n ============')
```

# 計算三種不同類型特徵的平均數、標準差

```python
print('Describe: \n', app_train[int_features].describe())
print('--------------------------')
print('Mean: \n', app_train[int_features].mean())
print('--------------------------')
print('STD: \n', app_train[int_features].std())
```

```python
print('Describe: \n', app_train[float_features].describe())
print('--------------------------')
print('Mean: \n', app_train[float_features].mean())
print('--------------------------')
print('STD: \n', app_train[float_features].std())
```

```python
print('Describe: \n', app_train[object_features].describe())
# print('--------------------------')
# print('Mean: \n', app_train[object_features].mean())
# print('--------------------------')
# print('STD: \n', app_train[object_features].std())
```

```python
type_data ={
    'Int_Features' : int_features,
    'Float_Features' : float_features,
    'Object_Fearutes' : object_features}

for d in type_data:
    print(f' {d} Describe: \n {app_train[type_data[d]].describe()} \n =====================' )
    if d == 'Object_Fearutes':
        break
    print(f' {d} Mean: \n {app_train[type_data[d]].mean()} \n =====================' )
    print(f' {d} STD: \n {app_train[type_data[d]].std()} \n =====================' )
```

# 畫出任意欄位的直方圖

```python
'By Pandas'
app_train['OWN_CAR_AGE'].plot.hist(alpha = 0.6)
```

```python
'By Matplotlib'
plt.hist(app_train['OWN_CAR_AGE'], alpha = 0.6)
plt.title('Own_Car_Age')
plt.xlabel('age')
plt.ylabel('Count')
plt.show()
```
